```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

### Database-Style DataFrame Joins

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 
                    'data1': range(7)})[['key', 'data1']]
df1
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'd'], 
                    'data2': range(3)})[['key', 'data2']]
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='key')
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                   'data1': range(7)})
df3
```

```python
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                   'data2': range(3)})
df4
```

```python
# Specify columns name if there's no any common column
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                    'data1': range(6)})
df1
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                    'data2': range(5)})
df2
```

```python
pd.merge(df1, df2, on='key', how='left')
```

```python
# Many-to-many joins returns cartesian product of rows
pd.merge(df1, df2, how='inner')
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3]})
left
```

```python
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})
right
```

```python
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```

### Merging on Index

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                     'value': range(6)})
left1
```

```python
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
lefth = pd.DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'], 
                      'key2': [2000, 2001, 2002, 2001, 2002],
                      'data': np.arange(5.)})
lefth
```

```python
righth = pd.DataFrame(np.arange(12).reshape((6, 2)),
                      index=[['Nevada', 'Nevada', 'Ohio', 'Ohio', 'Ohio', 'Ohio'], [2001, 2000, 2000, 2000, 2001, 2002]],
                      columns=['event1', 'event2'])
righth
```

```python
# Merging Hierarchical Indexed Data
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True)
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True, how='outer')
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                     index=['a', 'c', 'e'],
                     columns=['Ohio', 'Nevada'])
left2
```

```python
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=['b', 'c', 'd', 'e'],
                      columns=['Missouri', 'Alabama'])
right2
```

```python
# merging with indexes of both sides
pd.merge(left2, right2, how='outer', left_index=True, right_index=True)
```

```python
left2.join(right2, how='outer')
```

```python
left1.join(right1, on='key')
```

```python
another = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.],],
                      index=['a', 'c', 'e', 'f'],
                      columns=['New York', 'Oregon'])
```

```python
# index-on-index merges
left2.join([right2, another])
```

```python
left2.join([right2, another], how='outer')
```

### Concatenating Along an Axis

```python
arr = np.arange(12).reshape((3, 4))
arr
```

```python
np.concatenate([arr, arr], axis=1)
```

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2,3,4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
```

```python
pd.concat([s1, s2, s3])
```

```python
pd.concat([s1, s2, s3], axis=1)
```

```python
s4 = pd.concat([s1 * 5, s3])
```

```python
pd.concat([s1, s4], axis=1)
```

```python
pd.concat([s1, s4], axis=1, join='inner')
```

```python
pd.concat([s1, s4], axis=1, join_axes=[['a', 'c', 'b', 'e']])
```

```python
result = pd.concat([s1, s1, s3], keys=['one', 'two', 'three'])
result
```

```python
result.unstack()
```

```python
pd.concat([s1, s2, s3], axis=1, keys=['one', 'two', 'three'])
```

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'],
                   columns=['one', 'two'])
df1
```

```python
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2), index=['a', 'c'],
                   columns=['three', 'four'])
df2
```

```python
# same logic extends to DataFrame object
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'])
```

```python
# if a dict passed instead of a list, dict's keys will be used for the key option.
pd.concat({'level1': df1, 'level2': df2}, axis=1)
```

```python
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'], names=['upper', 'lower'])
```

```python
df1 = pd.DataFrame(np.random.randn(3, 4), columns=['a', 'b', 'c', 'd'])
df1
```

```python
df2 = pd.DataFrame(np.random.randn(2, 3), columns=['b', 'd', 'a'])
df2
```

```python
# pass ignore_index option if the row index does not contain any relevant data
pd.concat([df1, df2], ignore_index=True)
```

### Combining Data with Overlap

```python
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
             index=['f', 'e', 'd', 'c', 'b', 'a'])
a
```

```python
b = pd.Series(np.arange(len(a), dtype=np.float64),
             index=['f', 'e', 'd', 'c', 'b', 'a'])
b[-1] = np.nan
b
```

```python
np.where(pd.isnull(a), b, a)
```

```python
b[:-2].combine_first(a[2:])
```

```python
df1 = pd.DataFrame({'a': [1., np.nan, 5., np.nan],
                    'b': [np.nan, 2., np.nan, 6.],
                    'c': range(2, 18, 4)})
df1
```

```python
df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],
                    'b': [np.nan, 3., 4., 6., 8.]})
df2
```

```python
df1.combine_first(df2)
```
